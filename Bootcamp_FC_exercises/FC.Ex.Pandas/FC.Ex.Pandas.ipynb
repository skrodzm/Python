##### <font color='pink'>2.Po kolumnie z nazwiskiem dodaj kolumnę reprezentującą płeć danej osoby. Można wykorzystać fakt, że imiona kobiet kończą się literka 'a'.</font>

```python
df.insert(2, 'Płeć', '')
```

```python
df['Płeć'] = df['Imię'].apply(lambda x: 'F' if x.endswith('a') else 'M')
```

```python
df.head()
```

##### <font color='pink'>3.Posortuj zbiór danych alfabetycznie po nazwiskach, a w ramach jednego nazwiska po roku urodzenia.</font>

```python
df = df.sort_values(by=['Nazwisko', 'Rok urodzenia'])
```

```python
df.head()
```

##### <font color='pink'>4.Do zbioru danych należy dodać kolumnę, która będzie mówiła o poziomie doświadczenia danej osoby, w sposób słowny, przyjmując: Junior (do 3 lat doświadczenia), Mid (4 do 8 lat), Senior (9 do 12 lat), Ekspert (od 13 lat).</font>

```python
df.insert((df.columns.get_loc('Doświadczenie (w latach)') + 1), 'Doświadczenie (słownie)', '')
```

```python
def doswiadczenie(years):
    if years <= 3:
        return 'Junior'
    elif years >= 4 and years <= 8:
        return 'Mid'
    elif years >= 9 and years <= 12:
        return 'Senior'
    elif years >= 13:
        return 'Expert'
```

```python
df['Doświadczenie (słownie)'] = df['Doświadczenie (w latach)'].apply(doswiadczenie)
```

```python
df.head()
```

##### <font color='pink'>5.Kolumna reprezentująca rok dołączenia do firmy ma puste wartości. Uzupełnij brakujące wartości używając średniej dla osób o takim samym poziomie doświadczenia (kolumna z Zadania 3), w skali całej firmy.</font>

```python
df['Rok rozpoczęcia pracy w firmie'].isnull().sum()
```

```python
df['Rok rozpoczęcia pracy w firmie'] = df['Rok rozpoczęcia pracy w firmie'].fillna(df.groupby('Doświadczenie (słownie)')['Rok rozpoczęcia pracy w firmie'].transform('mean')).astype(int)
```

##### <font color='pink'> 7.Szef chciałby zmienić nazewnictwo departamentów w całej firmie. Napisz kod, który zamieni wartości w kolumnie reprezentującej departament, nastepująco: Software Development -> Software Development Team, HR -> Human Resources, Data Science -> Data Analytics.</font>
